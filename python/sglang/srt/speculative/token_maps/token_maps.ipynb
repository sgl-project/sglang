```python
def analyze_sequence(sequence):
    """
    Analyze a sequence to understand its properties and structure.
    
    Args:
        sequence: A list, tensor, or sequence-like object to analyze
    """
    # Verify what type of object this is
    print(f"Type: {type(sequence)}")
    print(f"Length: {len(sequence)}")
    print(f"First 10 elements: {sequence[:10]}")
    print(f"Last 10 elements: {sequence[-10:]}")

    # Check if all elements are integers
    print(f"All elements are integers: {all(isinstance(x, int) for x in sequence)}")

    # Get some statistics
    print(f"Min value: {min(sequence)}")
    print(f"Max value: {max(sequence)}")
    print(f"Number of unique values: {len(set(sequence))}")

    # Check if the sequence is contiguous
    is_contiguous = sequence == list(range(min(sequence), max(sequence) + 1))
    print(f"Is the sequence contiguous? {is_contiguous}")

    # Check if it's a contiguous sequence or has gaps
    sequence_set = set(sequence)
    expected_range = set(range(min(sequence), max(sequence) + 1))
    missing_values = expected_range - sequence_set
    print(f"Missing values in range: {len(missing_values)} out of {len(expected_range)}")
    if len(missing_values) <= 20:  # Only show if not too many
        print(f"Some missing values: {sorted(list(missing_values))[:20]}")

    # Check if the sequence is sorted
    is_sorted = sequence == sorted(sequence)
    print(f"Is the sequence sorted? {is_sorted}")

# Test the function with the original tensor
analyze_sequence(tokenizer_freq_hf)
```

## Prepare a token map from target var

```python
def get_token_map(token_scores: torch.Tensor, num_tokens_to_keep: int) -> torch.Tensor:
    """
    Find the indices of the tokens corresponding to maximal scores.
    """
    return token_scores.argsort(dim=-1, descending=True)[:num_tokens_to_keep]


target_var: torch.Tensor = torch.load("/home/projects/dharel/nadavt/repos/sglang/python/sglang/srt/speculative/token_maps/target-var-meta-llama-Llama-3.1-8B-Instruct-wikitext-wikitext-103-raw-v1-train.pt")
indices_to_keep: torch.Tensor = get_token_map(target_var, 10_000)
indices_to_keep

# save the indices to a file if not already there
filepath_indices_to_keep = "/home/projects/dharel/nadavt/repos/sglang/python/sglang/srt/speculative/token_maps/indices_to_keep.pt"
if not os.path.exists(filepath_indices_to_keep):
    torch.save(indices_to_keep, filepath_indices_to_keep)

# load the indices from the file
indices_to_keep_local = torch.load(filepath_indices_to_keep, map_location='cpu')
indices_to_keep_local

# check if the indices are the same
assert (indices_to_keep == indices_to_keep_local).all(), "The indices are not the same"
```
